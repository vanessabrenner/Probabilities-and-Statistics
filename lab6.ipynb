```python
from scipy.stats import norm
from numpy import mean, std, linspace
from matplotlib.pyplot import show, hist, grid, legend, xticks, plot

data = norm.rvs(loc=165,scale=10,size=10000)


hist(data, 16, density = True, rwidth = 0.9, color = 'magenta', edgecolor = 'black',
     alpha = 0.5, label = 'frecvente relative')
#x = range(130,210)
x = linspace(130,210,1000)
plot(x,norm.pdf(x,loc=165,scale=10),"-r")
xticks(range(130,210,10))
grid()
show()
```

```python
from scipy.stats import expon, uniform
from numpy import mean, std, multiply
from math import log

def alege():    
    u = uniform.rvs()        
    if(u < 0.4):
        return "I1"
    else:
        return "I2"

def exp(a):    
    x = (-1/a)*log(1 - uniform.rvs())
    return x

def timp():
    p = alege()
    if p == "I1":
        return exp(1/5)
    if p == "I2":
        return uniform.rvs(loc=4,scale=2)
data = []    

for i in range(1000):
    data.append(timp())

print(sum(data)/1000)
print(std(data))

#b)
nr = 0 
for i in range(1000):
    p = alege()
    prob = 0
    if p == "I1":
        prob = exp(1/5)
    if p == "I2":
        prob = uniform.rvs(loc=4,scale=2)

    if(prob < 5):
       nr += 1

print(nr/1000) 

#c)
rez = 0.4 * expon.cdf(5, loc = 0, scale = 5) + 0.6 * uniform.cdf(5, loc = 4, scale = 2)
print(rez)
```
