```python
from random import choices, sample
from math import comb, perm

#cel putin o bila extrasa este rosie
def A(lista):
    if "r" in lista:
        return True
    return False

#toate bilele extrase au aceeasi culoare
def B(lista):
    s = set(lista)
    if len(s) == 1:
        return True
    return False

def start():
    countA = 0
    countAB = 0
    for i in range(100000):
        l = sample(["r", "a", "v"], counts=[5,3,2], k=3)
        if A(l):
            countA+=1
            if B(l):
                countAB+=1
    return countAB/countA

def probTeoretic():
    probToateRosii = comb(5,3) / comb(10,3)
    probCelPutinUnaRosie = 1 - comb(3+2,3)/ comb(10,3)
    return probToateRosii / probCelPutinUnaRosie

start()
#probTeoretic()
```
