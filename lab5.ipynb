```python
from scipy.stats import uniform

distributie=[(0.1, "x1"),(0.3, "x2"),(0.2, "x3"),(0.2, "x4"),(0.1, "x5"),(0.1, "x6")]

u = uniform.rvs(size=1)
i = 0
rez = ""
for p in distributie:
    if(u > i):
        i = i + p[0]
        rez = p[1]
    else:
        break

print(u, ' ', rez)    
```

```python
from scipy.stats import uniform
from matplotlib.pyplot import hist
from matplotlib.pyplot import bar, hist, grid, show, legend

distributie=[(0.46, "0"),(0.4, "A"),(0.1, "B"),(0.04, "AB")]
data = []


for i in range(10000):
    u = uniform.rvs(size=1)
    r = 0 
    rez = ""
    for p in distributie:    
        if(u > r):
            r = r + p[0]
            rez = p[1]
        else:
            break
    data.append(rez)

bin_edges = [k+0.5 for k in range(-1,4)]
hist(data, bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black',
     alpha = 0.5, label = 'frecvente relative')
legend(loc = 'lower left')
```

```python
from scipy.stats import uniform, expon
from math import log
from matplotlib.pyplot import bar, show, hist, grid, legend, xticks , yticks, plot

def exp(a):    
    x = (-1/a)*log(1 - uniform.rvs())
    return x

data=[]
for i in range(1000):
    x = exp(1/12)
    data.append(x)

hist(data, 12, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black', range = (0,60),
     alpha = 0.5, label = 'frecvente relative')
legend(loc = 'lower left')

x = range(60)
plot(x,expon.pdf(x,loc=0,scale=14.5),'-r')
xticks(range(0,60,5))
grid()
show()
```
