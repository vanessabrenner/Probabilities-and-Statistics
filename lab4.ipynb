```python
from matplotlib.pyplot import hist
from scipy.stats import bernoulli, binom


data = []
for p in range(1000):
    poz = 0
    for i in range(10):
        pas = bernoulli.rvs(0.5)
        if pas == 1:
            poz += 1
        else:
            poz -= 1
        #print(poz)
        data.append(poz)

bin_edges = [k+0.5 for k in range(-11,11)]
hist(data, bin_edges, density = True, rwidth = 0.9, color = 'green', edgecolor = 'black',
     alpha = 0.5, label = 'frecvente relative')

```

```python
from scipy.stats import bernoulli
from random import randrange
from matplotlib.pyplot import bar, hist, grid, show, legend
poz = 0
rez = []
for e in range(10000):
    poz = 0
    for i in bernoulli.rvs(0.5, size=10):
        if (i == 1):
            poz = poz + 1
        else:
            poz = poz - 1
    rez.append(poz)
    poz = 0
    for i in bernoulli.rvs(0.5, size=11):
        if (i == 1):
            poz = poz + 1
        else:
            poz = poz - 1
    rez.append(poz % 5)
#print(rez)
bin_edges = [k+0.5 for k in range(-11,11)]
hist(rez, bin_edges, density = True, rwidth = 0.8, color = 'green', edgecolor = 'black',
alpha = 1, label = 'frecvente relative')
grid()
show()
```
