```python
from random import randint
from math import pow, perm
#from itertools import permutations 

def probability(n):
    nrCPosibile = pow(365,n)
    nrCFavorabile = perm(365,n)
    return 1 - nrCFavorabile / nrCPosibile

def prob():
    l = [randint(1, 365) for i in range(23)]
    s = set(l)
    if(len(l) != len(s)):
        return True
    return False

prob()
nr = 0
for i in range(0, 10000):
    if(prob()):
      nr += 1
print(nr / 10000)
probability(23)
```
